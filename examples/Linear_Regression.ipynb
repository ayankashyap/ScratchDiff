```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
```

```python
X,Y= make_regression(n_samples=200, n_features=1, noise=50,random_state=3)
plt.scatter(X[:, 0], Y , color= 'blue')
```

```python
def MSE(y,yhat):
    N = 1/y.data.shape[0]
    a = add(yhat , -y.data)
    s = sum(power(a, 2))
    loss =  mul(s, Unit(N, requires_grad=False))
    return loss
    
```

```python
X = Unit(X, requires_grad=False)
Y = Unit(Y.reshape(Y.shape[0], 1), requires_grad=False)
```

```python
print(X.shape)
print(Y.shape)
```

```python
b = Unit(np.zeros((X.shape[0],1)))
W = Unit(np.random.normal(size=(X.shape[1], 1)))
```

```python
plt.scatter(X.data[:, 0], X.data@W.data + b.data , color= 'red')
```

```python
for i in range(10000):
    W.zero_grad()
    b.zero_grad()
    yhat = add(matmul(X,W), b)
    loss = MSE(Y, yhat)
    print(loss)
    loss.backward_pass()
    W.data -= 0.001*W.grad.data
    b.data -= 0.001*b.grad.data
    
```

```python
y_true = [i[0] for i in Y.data]
y_pred = [i[0] for i in yhat.data]
m,y_int = np.polyfit(X.data[:, 0], y_pred, 1)
```

```python
#1000 epochs
line_vals = [m*x + y_int for x in X.data[:, 0]]
plt.scatter(X.data[:, 0], y_true, s=10, c='b', label='True')
plt.scatter(X.data[:, 0], y_pred,s=10, c='r', label='predicted')
plt.legend(loc='upper left');
```
